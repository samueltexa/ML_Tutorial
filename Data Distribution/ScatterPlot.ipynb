# A scatter plot is a diagram where each value in the data set is represented by a dot.

```python
import matplotlib.pyplot as plt
import pandas as np
import numpy as np
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6] #The x array represents the age of each car.

y = [99,86,87,88,111,86,103,87,94,78,77,85,86] # The y array represents the speed of each car.
```

```python
plt.scatter(x, y)
```

```python
# the two fastest cars were both 2 years old, and the slowest car was 12 years old.
```

# Random distribution

```python
x = np.random.normal(5.0, 1.0, 1000)
y = np.random.normal(10.0, 2.0, 1000)
plt.scatter(x, y)
```

```python
# the dots are concentrated around the value 5 on the x-axis, and 10 on the y-axis.
# the spread is wider on the y-axis than on the x-axis.
```
